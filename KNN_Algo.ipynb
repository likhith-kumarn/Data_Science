```python
# K-NN algorithm is used for classification and regression
# Classifying iris dataset using K-NN
# K-NN is Supervised ML
# It is used for Classification and Regression
```

```python
# label Encoder : Converts categorical data to numerical data
# LabelEncoder is data preprocessin technique
# It converst categorical data numerical data
# It returns numerical values as 0 and 1

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['variety']=le.fit_transform(df['variety'])
df
```

```python
df.head()
```

```python
df.describe()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df['variety']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
```

```python
# .fit() --> Learns from input data
# .fit_transform() ---> Learns and applies the parameters to model
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred1=model.predict(x_train)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
classification_report(y_test,y_pred)
```

```python
cm1=confusion_matrix(y_train,y_pred1)
cm1
```

```python
classification_report(y_train,y_pred1)
```
