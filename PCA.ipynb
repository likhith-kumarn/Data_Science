```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer(as_frame=True)
cancer
```

```python
x=df[cancer['feature_names']]
x.head()
```

```python
x_mean=x.mean()
x_mean
```

```python
x_std=x.std()
x_std
```

```python
z=(x-x_mean)/x_std
z
```

```python
c=z.cov()
c
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(c,annot=True)
plt.show()
```

```python
import numpy as np
eigenvalues,eigenvectors=np.linalg.eig(c)
eigenvalues
```

```python
eigenvalues.shape
```

```python
eigenvectors
```

```python
eigenvectors.shape
```

```python
explained_var=np.cumsum(eigenvalues)/np.sum(eigenvalues)
explained_var
```

```python
n_components=np.argmax(explained_var>0.50)+1
n_components
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=n_components)
pca.fit(z)
x_pca = pca.transform(z)
x_pca = pd.DataFrame(x_pca, columns=[f'PC{i+1}' for i in range(n_components)])
display(x_pca.head())
```

The `x_pca` DataFrame now contains the transformed data with the reduced number of principal components. This data can be used for further analysis, visualization, or as input for machine learning models.
